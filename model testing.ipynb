```python
import joblib

# Load the scaler
scaler = joblib.load('scaler.pkl')

# Load the trained model
model = joblib.load('ff_severity_model.pkl')
model_classification=joblib.load('ff_prediction_model.pkl')
```

```python
import numpy as np

# Example new test sample (same features as training data)
X_new = np.array([[31,0,82.6,5.8,3.1]])  # Replace with real test values

# Scale the input using the saved scaler
X_new_scaled = scaler.transform(X_new)
```

```python
# Make predictions
prediction_classification = model_classification.predict(X_new_scaled)
prediction_severity = model.predict(X_new_scaled)

print("Predicted Fire Classification:", prediction_classification)
print("Predicted Fire Severity:", prediction_severity)
```
